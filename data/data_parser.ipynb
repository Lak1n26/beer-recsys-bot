- –û—Ç–ª–∞–¥–æ—á–Ω—ã–π –≤—ã–≤–æ–¥ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å—Ç—Ä—É–∫—Ç—É—Ä—É –æ—Ç–≤–µ—Ç–∞ API
- –ö–æ–¥ –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ—Ç –æ–±–∞ —Ñ–æ—Ä–º–∞—Ç–∞: OpenAI-compatible –∏ YandexGPT Foundation Models


````python
import requests
import json
import time
from typing import Dict, List, Optional
import os
import urllib3

# –û—Ç–∫–ª—é—á–∞–µ–º –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏—è –æ –Ω–µ–ø—Ä–æ–≤–µ—Ä–µ–Ω–Ω—ã—Ö SSL —Å–µ—Ä—Ç–∏—Ñ–∏–∫–∞—Ç–∞—Ö
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# –û–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ: –∑–∞–≥—Ä—É–∑–∫–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –∏–∑ .env —Ñ–∞–π–ª–∞
# –†–∞—Å–∫–æ–º–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ, –µ—Å–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç–µ python-dotenv
# from dotenv import load_dotenv
# load_dotenv()

# –ö–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏—è API
with open('ELIZA_API_TOKEN.txt', 'r') as file:
    SOY_TOKEN = file.read().strip()
    
GPT_MODEL = "gpt-4o"

# URL –¥–ª—è YandexGPT API
GPT_URL = "https://api.eliza.yandex.net/openai/v1/chat/completions"

print("üîß –ö–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏—è –∑–∞–≥—Ä—É–∂–µ–Ω–∞")
if SOY_TOKEN:
    print("‚úì OAuth —Ç–æ–∫–µ–Ω –Ω–∞–π–¥–µ–Ω")
    print(f"üìù –ú–æ–¥–µ–ª—å: {GPT_MODEL}")
    print(f"üåê API endpoint: {GPT_URL}")
else:
    print("‚ö†Ô∏è  –í–Ω–∏–º–∞–Ω–∏–µ: OAuth —Ç–æ–∫–µ–Ω –Ω–µ –Ω–∞–π–¥–µ–Ω. –°–æ–∑–¥–∞–π—Ç–µ —Ñ–∞–π–ª ELIZA_API_TOKEN.txt")
````

````python
def call_yandex_gpt(prompt: str, temperature: float = 0.3) -> Optional[str]:
    """
    –í—ã–∑–æ–≤ YandexGPT API (ELIZA) –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞
    
    Args:
        prompt: –¢–µ–∫—Å—Ç –∑–∞–ø—Ä–æ—Å–∞ –∫ –º–æ–¥–µ–ª–∏
        temperature: –¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ (0.0 - –¥–µ—Ç–µ—Ä–º–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–π, 1.0 - –∫—Ä–µ–∞—Ç–∏–≤–Ω—ã–π)
    
    Returns:
        –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç –∏–ª–∏ None –≤ —Å–ª—É—á–∞–µ –æ—à–∏–±–∫–∏
    """
    headers = {
        "authorization": f"OAuth {SOY_TOKEN}",
        "content-type": "application/json"
    }
    
    payload = {
        "model": GPT_MODEL,
        "messages": [
            {
                "role": "system",
                "content": "–¢—ã —ç–∫—Å–ø–µ—Ä—Ç –ø–æ –ø–∏–≤—É –∏ —Å–æ–º–µ–ª—å–µ. –¢–≤–æ—è –∑–∞–¥–∞—á–∞ - –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª—è—Ç—å —Ç–æ—á–Ω—É—é –∏ —Å—Ç—Ä—É–∫—Ç—É—Ä–∏—Ä–æ–≤–∞–Ω–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Å–æ—Ä—Ç–∞—Ö –ø–∏–≤–∞."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        "temperature": temperature,
        "max_tokens": 2000
    }
    
    try:
        # –û—Ç–∫–ª—é—á–∞–µ–º –ø—Ä–æ–≤–µ—Ä–∫—É SSL —Å–µ—Ä—Ç–∏—Ñ–∏–∫–∞—Ç–∞ –¥–ª—è —Å–∞–º–æ–ø–æ–¥–ø–∏—Å–∞–Ω–Ω—ã—Ö —Å–µ—Ä—Ç–∏—Ñ–∏–∫–∞—Ç–æ–≤
        response = requests.post(GPT_URL, headers=headers, json=payload, timeout=30, verify=False)
        response.raise_for_status()
        
        result = response.json()
        
        # –§–æ—Ä–º–∞—Ç –æ—Ç–≤–µ—Ç–∞ ELIZA API
        if "response" in result and "choices" in result["response"]:
            choices = result["response"]["choices"]
            if len(choices) > 0 and "message" in choices[0] and "content" in choices[0]["message"]:
                return choices[0]["message"]["content"]
        
        # –ï—Å–ª–∏ —Ñ–æ—Ä–º–∞—Ç –Ω–µ–æ–∂–∏–¥–∞–Ω–Ω—ã–π, –≤—ã–≤–æ–¥–∏–º –æ—Ç–ª–∞–¥–æ—á–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
        print(f"‚ö†Ô∏è –ù–µ–æ–∂–∏–¥–∞–Ω–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç –æ—Ç–≤–µ—Ç–∞ API")
        print(f"üìã –î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–ª—é—á–∏: {list(result.keys())}")
        print(f"–ü–æ–ª–Ω—ã–π –æ—Ç–≤–µ—Ç:")
        print(json.dumps(result, indent=2, ensure_ascii=False))
        return None
            
    except requests.exceptions.RequestException as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã–∑–æ–≤–µ API: {e}")
        return None
    except Exception as e:
        print(f"–ù–µ–æ–∂–∏–¥–∞–Ω–Ω–∞—è –æ—à–∏–±–∫–∞: {e}")
        import traceback
        traceback.print_exc()
        return None
````

````python
def extract_beer_info(beer_name: str) -> Optional[Dict]:
    """
    –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –ø–∏–≤–µ —Å –ø–æ–º–æ—â—å—é YandexGPT
    
    Args:
        beer_name: –ù–∞–∑–≤–∞–Ω–∏–µ –ø–∏–≤–∞
    
    Returns:
        –°–ª–æ–≤–∞—Ä—å —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ –ø–∏–≤–µ –∏–ª–∏ None –≤ —Å–ª—É—á–∞–µ –æ—à–∏–±–∫–∏
    """
    prompt = f"""–ü—Ä–µ–¥–æ—Å—Ç–∞–≤—å –ø–æ–¥—Ä–æ–±–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø–∏–≤–µ "{beer_name}" –≤ —Ñ–æ—Ä–º–∞—Ç–µ JSON.

–°—Ç—Ä—É–∫—Ç—É—Ä–∞ JSON –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å —Å–ª–µ–¥—É—é—â–µ–π:
{{
    "name": "–Ω–∞–∑–≤–∞–Ω–∏–µ –ø–∏–≤–∞",
    "description": "–∫—Ä–∞—Ç–∫–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ –ø–∏–≤–∞ (–¥–æ 10 –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π)",
    "alcohol_percentage": "–ø—Ä–æ—Ü–µ–Ω—Ç –∞–ª–∫–æ–≥–æ–ª—è (—á–∏—Å–ª–æ –∏–ª–∏ –¥–∏–∞–ø–∞–∑–æ–Ω)",
    "bitterness": "—Å—Ç–µ–ø–µ–Ω—å –≥–æ—Ä–µ—á–∏ (IBU –∏–ª–∏ —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ: –Ω–∏–∑–∫–∞—è/—Å—Ä–µ–¥–Ω—è—è/–≤—ã—Å–æ–∫–∞—è)",
    "country": "—Å—Ç—Ä–∞–Ω–∞ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞",
    "taste_tags": ["—Ç–µ–≥1", "—Ç–µ–≥2", "—Ç–µ–≥3", ...],
    "beer_type": "—Ç–∏–ø –ø–∏–≤–∞ (—Å–≤–µ—Ç–ª–æ–µ/—Ç–µ–º–Ω–æ–µ/—è–Ω—Ç–∞—Ä–Ω–æ–µ/–¥—Ä—É–≥–æ–µ)",
    "style": "—Å—Ç–∏–ª—å –ø–∏–≤–∞ (IPA, –ª–∞–≥–µ—Ä, —ç–ª—å, —Å—Ç–∞—É—Ç –∏ —Ç.–¥.)"
}}

–ï—Å–ª–∏ –∫–∞–∫–∞—è-—Ç–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞, —É–∫–∞–∂–∏ "–Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ" –∏–ª–∏ null.
–í–µ—Ä–Ω–∏ —Ç–æ–ª—å–∫–æ JSON –±–µ–∑ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤."""

    response_text = call_yandex_gpt(prompt, temperature=0.2)
    
    if not response_text:
        return None
    
    try:
        # –ü–æ–ø—ã—Ç–∫–∞ –∏–∑–≤–ª–µ—á—å JSON –∏–∑ –æ—Ç–≤–µ—Ç–∞
        # –ò–Ω–æ–≥–¥–∞ –º–æ–¥–µ–ª—å –º–æ–∂–µ—Ç –¥–æ–±–∞–≤–∏—Ç—å —Ç–µ–∫—Å—Ç –¥–æ –∏–ª–∏ –ø–æ—Å–ª–µ JSON
        json_start = response_text.find('{')
        json_end = response_text.rfind('}') + 1
        
        if json_start != -1 and json_end > json_start:
            json_str = response_text[json_start:json_end]
            beer_info = json.loads(json_str)
            return beer_info
        else:
            print(f"–ù–µ —É–¥–∞–ª–æ—Å—å –Ω–∞–π—Ç–∏ JSON –≤ –æ—Ç–≤–µ—Ç–µ –¥–ª—è {beer_name}")
            print(f"–û—Ç–≤–µ—Ç: {response_text}")
            return None
            
    except json.JSONDecodeError as e:
        print(f"–û—à–∏–±–∫–∞ –ø–∞—Ä—Å–∏–Ω–≥–∞ JSON –¥–ª—è {beer_name}: {e}")
        print(f"–û—Ç–≤–µ—Ç: {response_text}")
        return None
    except Exception as e:
        print(f"–ù–µ–æ–∂–∏–¥–∞–Ω–Ω–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ {beer_name}: {e}")
        return None
````

````python
def collect_beer_data(
    beer_names: List[str], 
    output_file: str = "beer_data.json",
    delay: float = 1.0,
    resume_from: int = 0
) -> List[Dict]:
    """
    –°–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –æ –≤—Å–µ—Ö —Å–æ—Ä—Ç–∞—Ö –ø–∏–≤–∞ –∏–∑ —Å–ø–∏—Å–∫–∞
    
    Args:
        beer_names: –°–ø–∏—Å–æ–∫ –Ω–∞–∑–≤–∞–Ω–∏–π –ø–∏–≤–∞
        output_file: –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
        delay: –ó–∞–¥–µ—Ä–∂–∫–∞ –º–µ–∂–¥—É –∑–∞–ø—Ä–æ—Å–∞–º–∏ –≤ —Å–µ–∫—É–Ω–¥–∞—Ö (–¥–ª—è –∏–∑–±–µ–∂–∞–Ω–∏—è rate limits)
        resume_from: –ò–Ω–¥–µ–∫—Å, —Å –∫–æ—Ç–æ—Ä–æ–≥–æ –ø—Ä–æ–¥–æ–ª–∂–∏—Ç—å —Å–±–æ—Ä (–¥–ª—è –≤–æ–∑–æ–±–Ω–æ–≤–ª–µ–Ω–∏—è –ø–æ—Å–ª–µ –ø—Ä–µ—Ä—ã–≤–∞–Ω–∏—è)
    
    Returns:
        –°–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ –ø–∏–≤–µ
    """
    all_beer_data = []
    
    # –ü–æ–ø—ã—Ç–∫–∞ –∑–∞–≥—Ä—É–∑–∏—Ç—å —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ
    if resume_from > 0 and os.path.exists(output_file):
        try:
            with open(output_file, 'r', encoding='utf-8') as f:
                all_beer_data = json.load(f)
            print(f"–ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(all_beer_data)} –∑–∞–ø–∏—Å–µ–π –∏–∑ —Å—É—â–µ—Å—Ç–≤—É—é—â–µ–≥–æ —Ñ–∞–π–ª–∞")
        except Exception as e:
            print(f"–ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ: {e}")
    
    total = len(beer_names)
    
    for i, beer_name in enumerate(beer_names[resume_from:], start=resume_from):
        print(f"\n[{i+1}/{total}] –û–±—Ä–∞–±–æ—Ç–∫–∞: {beer_name}")
        
        beer_info = extract_beer_info(beer_name)
        
        if beer_info:
            all_beer_data.append(beer_info)
            print(f"‚úì –£—Å–ø–µ—à–Ω–æ —Å–æ–±—Ä–∞–Ω–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ {beer_name}")
        else:
            print(f"‚úó –ù–µ —É–¥–∞–ª–æ—Å—å —Å–æ–±—Ä–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ {beer_name}")
            # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ö–æ—Ç—è –±—ã –Ω–∞–∑–≤–∞–Ω–∏–µ
            all_beer_data.append({
                "name": beer_name,
                "error": "–ù–µ —É–¥–∞–ª–æ—Å—å –ø–æ–ª—É—á–∏—Ç—å –¥–∞–Ω–Ω—ã–µ"
            })
        
        # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∫–∞–∂–¥—ã–µ 10 –∑–∞–ø–∏—Å–µ–π
        if (i + 1) % 10 == 0:
            try:
                with open(output_file, 'w', encoding='utf-8') as f:
                    json.dump(all_beer_data, f, ensure_ascii=False, indent=2)
                print(f"üíæ –ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ: {len(all_beer_data)} –∑–∞–ø–∏—Å–µ–π")
            except Exception as e:
                print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–∏: {e}")
        
        # –ó–∞–¥–µ—Ä–∂–∫–∞ –º–µ–∂–¥—É –∑–∞–ø—Ä–æ—Å–∞–º–∏
        if i < total - 1:
            time.sleep(delay)
    
    # –§–∏–Ω–∞–ª—å–Ω–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ
    try:
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(all_beer_data, f, ensure_ascii=False, indent=2)
        print(f"\n‚úì –í—Å–µ –¥–∞–Ω–Ω—ã–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ {output_file}")
        print(f"–í—Å–µ–≥–æ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–æ: {len(all_beer_data)} —Å–æ—Ä—Ç–æ–≤ –ø–∏–≤–∞")
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–º —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–∏: {e}")
    
    return all_beer_data
````
